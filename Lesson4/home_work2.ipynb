```python
t = np.linspace(0, 10, 51)
f = np.cos(t)
```

```python
plt.plot(t, f, color='green')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim(-5, 5)

ax2.plot(x, y2)
ax2.set_title('График y2')

ax3.plot(x, y3)
ax3.set_title('График y3')

ax4.plot(x, y4)
ax4.set_title('График y4')

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

plt.show()
```

```python
plt.style.use('fivethirtyeight')

df = pd.read_csv('creditcard.csv')
```

```python
df.Class.value_counts().plot(kind='bar')
plt.show()
```

```python
df.Class.value_counts().plot(kind='bar', logy=True)
plt.show()
```

```python
plt.hist(df['V1'].loc[df.Class == 0], bins=20, density=True, alpha=0.5, color='grey', label='Class 0')
plt.hist(df['V1'].loc[df.Class == 1], bins=20, density=True, alpha=0.5, color='red', label='Class 1')
plt.legend(loc='upper left')
plt.xlabel('Class')
plt.show()
```
